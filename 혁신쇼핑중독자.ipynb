### 성별 : 상품명에 성별을 나타내는 단어 존재

> 주로 '속옷', '의류', '이미용' 상품군에서 발견됨

```python
def make_gender(row):
  if '여성' in row['상품명']:
    return '여성'
  elif '남성' in row['상품명']:
    return '남성'
  else:
    return '남/여 공용'
data['성별'] = data.apply(lambda x: make_gender(x), axis=1)
data
```

> 직접적인 성별 언급 외에도 성별을 암시하는 단어 존재

```python
# 보완
def modify_gender(row):
  name = row['상품명']
  # name = row['찐상품명']
  if ('드로즈' in name) | ('트렁크' in name):
    return '남성'
  elif '브라' in name:
    if row['상품군'] == '속옷':
      return '여성'
    else:
      return '남/여 공용'
  elif ('치마' in name) | ('언더탑' in name):
      return '여성'
  else:
    return row['성별']

data['성별'] = data.apply(lambda x: modify_gender(x), axis=1)
data
```

### 시간대 & 식사시간

- 시간대 (대략적으로) : 아침 (06:00 ~ 11:00) / 점심 (11:00 ~ 16:00) / 저녁 (16:00 ~ 21:00) / 새벽 (21:00 ~ 26:00)
- 식사시간 (아침, 점심, 저녁 각 1시간 30분 씩) : 식사시간 or 비식사시간

```python
def time_period(row):
  time = str(row['방송일시'])[-8:]
  if (time >='06') & (time <'11'):
    return '아침'
  elif (time >='11') & (time <'16'):
    return '점심'
  elif (time >='16') & (time <'21'):
    return '저녁'
  elif (time >='21') | (time <'3'):
    return '심야'

def meal_time(row):
  time = str(row['방송일시'])[-8:]
  if (time >='07:30') & (time <'09'):     # 아침
    return '식사시간'
  elif (time >='11:30') & (time <'13'):   # 점심
    return '식사시간'
  elif (time >='18') & (time <'19:30'):   # 저녁
    return '식사시간'
  else:
    return '비식사시간'
# df[(df['시간대'] >'02:00:00')&(df['시간대'] <'06')]   # 새벽 2시 ~ 6시 방송 없음. (단, 2시 10분 스케쥴 2개 있음.)
data['시간대'] = data.apply(lambda x: time_period(x), axis=1)
data['식사시간'] = data.apply(lambda x: meal_time(x), axis=1)
data
```

### 요일

```python
def get_day(row):
    time = row['datetime']
    if time.strftime("%A") == "Monday":
        return 'mon'
    elif time.strftime("%A") == "Tuesday":
        return 'tue'
    elif time.strftime("%A") == "Wednesday":
        return 'wed'
    elif time.strftime("%A") == "Thursday":
        return 'thu'
    elif time.strftime("%A") == "Friday":
        return 'fri'
    elif time.strftime("%A") == "Saturday":
        return 'sat'
    elif time.strftime("%A") == "Sunday":
        return 'sun'
data['datetime'] = pd.to_datetime(data['방송일시'])
data['요일'] = data.apply(lambda x: get_day(x), axis=1)
data = data.drop('datetime', axis=1)
data
```

##### 찐상품명 생성 (상품명에서 무이자, 일시불 제거)

```python
p = re.compile('$무\)')
q = re.compile('$유\)')

def real_name(row):
    name = row['상품명']
    if row['무/일'] == '구분x':
        return name
    elif row['무/일'] == '무이자':
        if '무이자' in name:
            return name[:name.index('무이자')] + name[name.index('무이자')+3:].strip()
        elif str('(무)') in name:
            return name[:name.index(str('(무)'))] + name[name.index(str('(무)'))+3:].strip()
        elif p.search(name):
            print("무", p.findall(name))
    elif row['무/일'] == '일시불':
        if '일시불' in name:
            return name[:name.index('일시불')] + name[name.index('일시불')+3:].strip()
        elif str('(일)') in name:
            return name[:name.index(str('(일)'))] + name[name.index(str('(일)'))+3:].strip()
        elif q.search(name):
            print("일", q.findall(name))
    
data['찐상품명'] = data.apply(lambda x: real_name(x), axis=1)
data['찐상품명'].nunique(), data['찐상품명'].unique()
```

##### 불필요한 괄호, 따옴표 제거

```python
def find_bracket(row):
  name = row['찐상품명']
  
  result = ''
  brackets = ['[', ']', '(', ')', '{', '}', '\"', "\'"]
  for b in brackets:
    if b in name:
      split = name.split(b)
      # 괄호가 마지막 단어에 들어 있을 경우
      if len(split[-1].split(' ')) == 1:
        if (b == '[') & (']' in split[-1]):   # 그 안에 닫히는 괄호가 또 있으면, 그 괄호만 삭제
          split[-1] = ' '.join(split[-1].split(']')[1:])
        elif (b == '(') & (')' in split[-1]):
          split[-1] = ' '.join(split[-1].split(')')[1:])
        elif (b == '{') & ('}' in split[-1]):
          split[-1] = ' '.join(split[-1].split('}')[1:])
        else:
          split = split[:-1]    # 아니면 (ex. 오타) 그냥 뒤에 다 날림
      if len(split) > 1:
        for n, s in enumerate(split):
          split[n] = s.strip()
      name = ' '.join(split)
  return name

data['찐상품명'] = data.apply(lambda x: find_bracket(x),axis=1)
data['찐상품명'].nunique(), data['찐상품명'].unique()
```

##### "+" 제거

```python
def get_plus(row):
  name = row['찐상품명']
  if '1+1' in name:
    plus = '1+1'
    split = name.split(plus)
    for n, s in enumerate(split):
      split[n] = s.strip()
    name = ' '.join(split)
  if '+' in name:
    plus = '+'
    split = name.split(plus)
    for n, s in enumerate(split):
      split[n] = s.strip()
    name = ' '.join(split)
  return name

# data23['찐상품명'] = data23.apply(lambda x: get_plus(x), axis=1)
data['찐상품명'] = data.apply(lambda x: get_plus(x), axis=1)
data['찐상품명'].nunique(), data['찐상품명'].unique()
```

##### n종, n팩 꼴의 단어 (n은 자연수)

```python
import re
p = re.compile('[0-9]+종')
q = re.compile('[0-9]+팩')

def get_jong(row):
    name = row['찐상품명']
    f = p.findall(name)
    if len(f) == 1:
        s = p.search(name)
        if s:
            start, end = s.span()
            name = name[:start].strip() + ' ' + name[end:].strip()
    elif len(f) > 1:
        for a in f:
            idx = name.find(a)
            name = name[:idx].strip() + ' ' + name[idx+len(a):].strip()
        
    f = q.findall(name)
    if len(f) == 1:
        s = q.search(name)
        if s:
            start, end = s.span()
            name = name[:start].strip() + ' ' + name[end:].strip()
    elif len(f) > 1:
        for a in f:
            idx = name.find(a)
            name = name[:idx].strip() + ' ' + name[idx+len(a):].strip()
    return name
# len(list(data23[data23['찐상품명'].str.contains(r'[0-9]+종')]['찐상품명'])), list(data23[data23['찐상품명'].str.contains(r' [0-9]+종')]['찐상품명'])
data['찐상품명'] = data.apply(lambda x: get_jong(x), axis=1)
data['찐상품명'].nunique(), data['찐상품명'].unique()
```

##### 세트, 시즌, 패키지, n차 제거

```python
import re
p = re.compile('[0-9]+세트')
q = re.compile('[0-9]+차')
r = re.compile('시즌[0-9]+')

def clean(row):
  # regex로 먼저 n세트
  name = row['찐상품명']
  f = p.findall(name)
  if len(f) == 1:
    s = p.search(name)
    if s:
      start, end = s.span()
      name = name[:start].strip() + ' ' + name[end:].strip()
  elif len(f) > 1:
    for a in f:
      idx = name.find(a)
      name = name[:idx].strip() + ' ' + name[idx+len(a):].strip()

  # regex로 먼저 시즌n
  f = r.findall(name)
  if len(f) == 1:
    s = r.search(name)
    if s:
      start, end = s.span()
      name = name[:start].strip() + ' ' + name[end:].strip()
  elif len(f) > 1:
    for a in f:
      idx = name.find(a)
      name = name[:idx].strip() + ' ' + name[idx+len(a):].strip()

  # 패키지
  pkg_split = name.split("패키지")
  if len(pkg_split) >= 2:
    for n, s in enumerate(pkg_split):
      pkg_split[n] = s.strip()
    name = ' '.join(pkg_split)
  
  # 세트
  set_split = name.split("세트")
  if len(set_split) >= 2:
    for n, s in enumerate(set_split):
      set_split[n] = s.strip()
    name = ' '.join(set_split)

  # n차
  f = q.findall(name)
  if len(f) == 1:
    s = q.search(name)
    if s:
      start, end = s.span()
      name = name[:start].strip() + ' ' + name[end:].strip()
  elif len(f) > 1:
    for a in f:
      idx = name.find(a)
      name = name[:idx].strip() + ' ' + name[idx+len(a):].strip()
  
  # 시즌
  season_split = name.split("시즌")
  if len(season_split) >= 2:
    space_split = name.split(' ')
    if '시즌' in space_split[-1]:
      season_split = space_split[-1].split('시즌')
      for n, s in enumerate(season_split):
        season_split[n] = s.strip()
      space_split[-1] = ' '.join(season_split)
    for n, s in enumerate(space_split):
      space_split[n] = s.strip()
    name = ' '.join(space_split)

  return name

data['찐상품명'] = data.apply(lambda x: clean(x),axis=1)
data['찐상품명'].nunique(), data['찐상품명'].unique()
```

### 어제상품군

### D-절기

다음 절기까지 남은 날짜 수

```python
import datetime

def cal_jeolgi(x):
  bangsongil = str(x['방송일시']).split(' ')[0]
  if x['절기여부'] == '절기':
    # print(0)
    return 0
  else:
    next_jeolgi = ''
    for j in jeolgi:
      if bangsongil < j:
        next_jeolgi = j
        break
    next_jeolgi = datetime.datetime.strptime(next_jeolgi, "%Y-%m-%d")
    today = datetime.datetime.strptime(bangsongil, "%Y-%m-%d")
    return int(str(next_jeolgi - today).split(',')[0].split(' ')[0])
    

data['D-절기'] = data.apply(lambda x: cal_jeolgi(x), axis=1)
data
```

### 마더코드 & 상품코드

마더코드와 상품코드는 무질서한 랜덤 코드로 각 코드들은 (기준을 알 수 없는) 특정 범위의 상품들에 매핑되어 있다.  
2019년 데이터와 2020년 6월 데이터의 마더코드, 상품코드 분포를 비교하여 본 결과, 겹치는 결과도 없을 뿐더러, 각각의 분포가 연관성이 거의 없어 보였기에 예측에 큰 효과를 보이지 않을 것으로 예상했고, 따라서 학습데이터인 2019년 데이터에서 2020년 6월 데이터에 존재하는 마더코드, 상품코드만 남기고 다 0 처리하기로 하였다.

```python
m19 = data_2019['마더코드'].apply(lambda x: x if x in data_2020['마더코드'].unique() else 0)
m20 = data_2020['마더코드'].apply(lambda x: x if x in data_2019['마더코드'].unique() else 0)
data['겹치는마더코드'] = pd.concat([m19, m20])
s19 = data_2019['상품코드'].apply(lambda x: x if x in data_2020['상품코드'].unique() else 0)
s20 = data_2020['상품코드'].apply(lambda x: x if x in data_2019['상품코드'].unique() else 0)
data['겹치는상품코드'] = pd.concat([s19, s20])
data
```
